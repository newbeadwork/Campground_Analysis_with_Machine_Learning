```python
import pandas as pd
```

```python
#The goal of this file is to create a dataset with all campgrounds, number of campsites and activities for each campground
#across US
```

```python
#Reading the data with campsites in order to count number of camsites for each campground
data_campsites_info = pd.read_csv("../../Resources/data/Campsites_API_v1.csv", encoding= 'unicode_escape')
data_campsites_info
```

```python
#As we are interested in overnight camping checking only this type
data_campsites_info["TypeOfUse"].unique()
```

```python
#Extracting the data for overnight nad multiuse campsites, excluding day use
data_campsites_overnight = data_campsites_info.loc[data_campsites_info["TypeOfUse"] != 'Day']
data_campsites_overnight
```

```python
#Checking number of unique campgrounds in the dataset
data_campsites_overnight["FacilityID"].nunique()
```

```python
#Counting number of campsites for each campground
number_of_campsites = data_campsites_overnight.groupby("FacilityID")["CampsiteID"].count()
number_of_campsites
```

```python
#Creating a dataframe with number of campsites
number_of_campsites = pd.DataFrame(number_of_campsites)
number_of_campsites.reset_index(inplace=True)
number_of_campsites
```

```python
#Renaming a column
number_of_campsites.rename(columns={"CampsiteID": "NumberofCampsites"}, inplace=True)
```

```python
#Checking data types
number_of_campsites.dtypes
```

```python
#Reading a dataset which describes campgrounds (facilities)
data_facilities = pd.read_csv("../../Resources/data/Facilities_API_v1.csv")
data_facilities
```

```python
#Checking columns 
data_facilities.columns
```

```python
#Finding facility types description as we interested in campgrounds only
data_facilities["FacilityTypeDescription"].unique()
```

```python
#Extracting the data only for campgrounds
data_campgrounds = data_facilities.loc[(data_facilities["FacilityTypeDescription"] == "Campground") |
                                     (data_facilities["FacilityTypeDescription"] == "Facility") |
                                      (data_facilities["FacilityTypeDescription"] == "Federal")]
data_campgrounds
```

```python
data_campgrounds["Reservable"].unique()
```

```python
#As we are interested only in reservable campgrounds extracting this data
data_campgrounds = data_campgrounds.loc[data_campgrounds["Reservable"] == True]
data_campgrounds
```

```python
#Dropping unnecessary columns 
data_campgrounds = data_campgrounds.drop(columns = ['ParentOrgID',
       'FacilityDescription', 'FacilityTypeDescription', 'FacilityUseFeeDescription',
       'FacilityDirections', 'FacilityPhone', 'FacilityEmail', 'FacilityReservationURL', 'FacilityMapURL', 
       'StayLimit', 'Reservable', 'Enabled', 'LastUpdatedDate'], axis = 1)
data_campgrounds
```

```python
#Checking datatypes
data_campgrounds.dtypes
```

```python
#Translating "FacilityID" to another datatype in order to merge both dataframes
data_campgrounds['FacilityID'] = data_campgrounds['FacilityID'].apply(pd.to_numeric) 
```

```python
#Checking transformation
data_campgrounds.dtypes
```

```python
#Creating united datasets with number of campsites for each campground
data_campgrounds_campsites = data_campgrounds.merge(number_of_campsites, on="FacilityID", how='inner')
data_campgrounds_campsites
```

```python
#Reading the data about activities for each facility (including campgrounds)
data_activities_facilities = pd.read_csv("../../Resources/Data/EntityActivities_API_v1.csv")
data_activities_facilities
```

```python
#Checking column "EntityType" in order to find only campgrounds
data_activities_facilities["EntityType"].unique()
```

```python
#Extracting the data only for campgrounds
data_activities_campgrounds = data_activities_facilities.loc[(data_activities_facilities["EntityType"] == "Campground") |
                                                             (data_activities_facilities["EntityType"] == "Facility") |
                                                             (data_activities_facilities["EntityType"] == "Federal")]
data_activities_campgrounds
```

```python
#Dropping unnecessary columns
data_activities_campgrounds = data_activities_campgrounds.drop(columns=["ActivityDescription", "ActivityFeeDescription"], 
                                                               axis = 1)
data_activities_campgrounds
```

```python
#Reading the data about activities names in order to add them to full campgrounds dataset
activities_names = pd.read_csv("../../Resources/data/Activities_API_v1.csv")
activities_names
```

```python
#Dropping unnecessary columns
activities_names = activities_names.drop(columns=["ActivityParentID", "ActivityLevel"])
activities_names
```

```python
#Creating dataframe in order to connect activities names and campgrounds
activities_names_campgrounds = data_activities_campgrounds.merge(activities_names, on="ActivityID", how="inner")
activities_names_campgrounds
```

```python
#"Packing" all activities names for each campground for more sufficient dataset
activities_grouped = activities_names_campgrounds.groupby("EntityID", as_index=False).agg({"ActivityName" : ",".join})  
activities_grouped = activities_grouped.rename(columns={"EntityID": "FacilityID"})
activities_grouped
```

```python
#Creating a full dataset with campgrounds, number of campsites and activities across USA
campgrounds_campsites_activities = data_campgrounds_campsites.merge(activities_grouped, on="FacilityID", how='left')
campgrounds_campsites_activities
```

```python
#Dropping unnecessary columns
campgrounds_campsites_activities = campgrounds_campsites_activities.drop(columns=["LegacyFacilityID",
                                                                                "FacilityAdaAccess",
                                                                                "Keywords"])
```

```python
#Storing the data inti csv
campgrounds_campsites_activities.to_csv("../../Resources/data/US_campgrounds_campsites_activities.csv", 
                                        index = False, header=True)
```
