```python
import pandas as pd
```

```python
data_facilities = pd.read_csv("full_data_activities_facilities.csv", encoding= 'unicode_escape')
```

```python
data_West_Coast = pd.read_csv("reservations_West_Coast.csv", encoding= 'unicode_escape')
```

```python
data_West_Coast.columns
```

```python
number_of_reservations = data_West_Coast.groupby("facilityid").size()
number_of_reservations
```

```python
number_of_reservations = pd.DataFrame(number_of_reservations)
number_of_reservations
```

```python
number_of_reservations.reset_index(inplace=True)
number_of_reservations = number_of_reservations.rename(columns={"facilityid": "FacilityID",
                                                                0: "NumberofReservations"})
number_of_reservations
```

```python
number_of_reservations["FacilityID"].nunique()
```

```python
data_facilities_reservations_West_Coast = data_facilities.merge(number_of_reservations, on="FacilityID", how='inner')
data_facilities_reservations_West_Coast
```

```python
facilities_adress = pd.read_csv("RIDBFullExport_V1_CSV/FacilityAddresses_API_v1.csv", encoding= 'unicode_escape')
```

```python
facilities_adress.columns
```

```python
facilities_state = facilities_adress.drop(columns=['FacilityAddressID', 'FacilityAddressType',
       'FacilityStreetAddress1', 'FacilityStreetAddress2',
       'FacilityStreetAddress3', 'City', 
       'AddressCountryCode', 'LastUpdatedDate'])
facilities_state
```

```python
data_facilities_reservations_West_Coast = data_facilities_reservations_West_Coast.merge(facilities_state, on="FacilityID", how='inner')
data_facilities_reservations_West_Coast
```

```python
data_facilities_reservations_West_Coast.groupby("PostalCode").size()
```

```python
data_facilities_reservations_West_Coast["FacilityID"].nunique()
```

```python
data_facilities_reservations_West_Coast["PostalCode"].mask(data_facilities_reservations_West_Coast["PostalCode"] == "California", "CA", inplace=True)

data_facilities_reservations_West_Coast["PostalCode"].mask(data_facilities_reservations_West_Coast["PostalCode"] == "OREGON", "OR", inplace=True)

data_facilities_reservations_West_Coast["PostalCode"].mask(data_facilities_reservations_West_Coast["PostalCode"] == "Oregon", "OR", inplace=True)

data_facilities_reservations_West_Coast["PostalCode"].mask(data_facilities_reservations_West_Coast["PostalCode"] == "WASHINGTON", "WA", inplace=True)

data_facilities_reservations_West_Coast["PostalCode"].mask(data_facilities_reservations_West_Coast["PostalCode"] == "Washington", "WA", inplace=True)

```

```python
data_facilities_reservations_West_Coast.groupby("PostalCode").size()
```

```python
data_facilities_reservations_West_Coast = data_facilities_reservations_West_Coast.loc[
                                            (data_facilities_reservations_West_Coast["PostalCode"] == "CA") |
                                            (data_facilities_reservations_West_Coast["PostalCode"] == "OR") |
                                            (data_facilities_reservations_West_Coast["PostalCode"] == "WA") 
]
data_facilities_reservations_West_Coast
```

```python
data_facilities_reservations_West_Coast["FacilityName"].nunique()
```

```python
duplicate = data_facilities_reservations_West_Coast[data_facilities_reservations_West_Coast.duplicated(subset=["FacilityName"], keep=False)]
duplicate
```

```python
data_facilities_reservations_West_Coast = data_facilities_reservations_West_Coast.drop_duplicates(subset=['FacilityID'])
data_facilities_reservations_West_Coast
```

```python
data_facilities_reservations_West_Coast.to_csv()
```
