```python
import pandas as pd
```

```python
data_campsites_info = pd.read_csv("RIDBFullExport_V1_CSV/Campsites_API_v1.csv")
data_campsites_info
```

```python
data_campsites_info["CampsiteType"].unique()
```

```python
data_campsites_info["TypeOfUse"].unique()
```

```python
data_campsites_info = data_campsites_info.loc[data_campsites_info["TypeOfUse"] != 'Day']
data_campsites_info
```

```python
data_campsites_info["FacilityID"].nunique()
```

```python
number_of_campsites = data_campsites_info.groupby("FacilityID")["CampsiteID"].count()
number_of_campsites
```

```python
number_of_campsites = pd.DataFrame(number_of_campsites)
number_of_campsites
```

```python
number_of_campsites.reset_index(inplace=True)
number_of_campsites
```

```python
number_of_campsites.rename(columns={"CampsiteID": "NumberofCampsites"}, inplace=True)
```

```python
number_of_campsites.dtypes
```

```python
data_facilities = pd.read_csv("RIDBFullExport_V1_CSV/Facilities_API_v1.csv")
data_facilities
```

```python
data_facilities.columns
```

```python
data_facilities["FacilityTypeDescription"].unique()
```

```python
data_facilities = data_facilities.loc[(data_facilities["FacilityTypeDescription"] == "Campground") |
                                     (data_facilities["FacilityTypeDescription"] == "Facility") |
                                      (data_facilities["FacilityTypeDescription"] == "Federal")]
data_facilities
```

```python
data_facilities["Reservable"].unique()
```

```python
data_facilities = data_facilities.loc[data_facilities["Reservable"] == True]
data_facilities
```

```python
data_facilities = data_facilities.drop(columns = ['ParentOrgID',
       'FacilityDescription', 'FacilityTypeDescription', 'FacilityUseFeeDescription',
       'FacilityDirections', 'FacilityPhone', 'FacilityEmail', 'FacilityReservationURL', 'FacilityMapURL', 
       'StayLimit', 'Reservable', 'Enabled', 'LastUpdatedDate'], axis = 1)
data_facilities
```

```python
data_facilities["Keywords"].nunique()
```

```python
data_facilities.dtypes
```

```python
data_facilities[['FacilityID', 'LegacyFacilityID']] = data_facilities[['FacilityID', 'LegacyFacilityID']].apply(pd.to_numeric) 
```

```python
data_facilities.dtypes
```

```python
data_facilities_campsites = data_facilities.merge(number_of_campsites, on="FacilityID", how='inner')

data_facilities_campsites
```

```python
data_facilities_campsites["NumberofCampsites"].max()
```

```python
data_activities_facilities = pd.read_csv("RIDBFullExport_V1_CSV/EntityActivities_API_v1.csv")
data_activities_facilities
```

```python
data_activities_facilities["EntityType"].unique()
```

```python
data_activities_facilities = data_activities_facilities.loc[(data_activities_facilities["EntityType"] == "Campground") |
                                     (data_activities_facilities["EntityType"] == "Facility") |
                                      (data_activities_facilities["EntityType"] == "Federal")]
data_activities_facilities
```

```python
data_activities_clean = data_activities_facilities.drop(columns=["ActivityDescription", "ActivityFeeDescription"], axis = 1)
data_activities_clean
```

```python
activities_description = pd.read_csv("RIDBFullExport_V1_CSV/Activities_API_v1.csv")
activities_description
```

```python
activities_description = activities_description.drop(columns=["ActivityParentID", "ActivityLevel"])
activities_description
```

```python
activities_facilities = data_activities_clean.merge(activities_description, on="ActivityID", how="inner")
activities_facilities
```

```python
activities_grouped = activities_facilities.groupby("EntityID", as_index=False).agg({"ActivityName" : ",".join})  
activities_grouped = activities_grouped.rename(columns={"EntityID": "FacilityID"})
activities_grouped
```

```python
full_data_activities_facilities = data_facilities_campsites.merge(activities_grouped, on="FacilityID", how='left')
full_data_activities_facilities
```

```python
full_data_activities_facilities = full_data_activities_facilities.drop(columns=["LegacyFacilityID",
                                                                                "FacilityAdaAccess",
                                                                                "Keywords"])
```

```python
full_data_activities_facilities.to_csv("full_data_activities_facilities.csv", index = False, header=True)
```
