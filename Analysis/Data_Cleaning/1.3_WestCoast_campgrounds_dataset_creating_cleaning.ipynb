```python
import pandas as pd
```

```python
#The goal of this file is to create one dataset of West Coast campgrounds with:
# - campgrounds name, location;
# - number of campsites;
# - campground activities;
# - number of reservations for 2016-2020
```

```python
#Reading the data of reservations for West Coast campgrounds for 2016-2020 (from step 1.1).
#!Due to a large size of the dataset (exceeds limits of GitHub repo) cleaning and merging was done on a local computer
data_West_Coast = pd.read_csv("../../Resources/data/reservations_West_Coast.csv", encoding= 'unicode_escape')
```

```python
#Reading the data of US federal reservable campgrounds (from step 1.2)
data_US_campgrounds = pd.read_csv("../../Resources/data/US_campgrounds_campsites_activities.csv", encoding= 'unicode_escape')
```

```python
#Checking columns
data_West_Coast.columns
```

```python
#Counting the number of reservations 2016-2020 for each West Coast campground
number_of_reservations = data_West_Coast.groupby("facilityid").size()
number_of_reservations
```

```python
#Creating a dataframe for future merging with US campgrounds dataset
number_of_reservations = pd.DataFrame(number_of_reservations)
number_of_reservations.reset_index(inplace=True)
number_of_reservations = number_of_reservations.rename(columns={"facilityid": "FacilityID",
                                                                0: "NumberofReservations"})
number_of_reservations
```

```python
#Checking the number of unique campgrounds
number_of_reservations["FacilityID"].nunique()
```

```python
#Merging
data_West_Coast_campgrounds = data_US_campgrounds.merge(number_of_reservations, on="FacilityID", how='inner')
data_West_Coast_campgrounds
```

```python
#Reading the data about campgrounds address for future merging
campgrounds_address = pd.read_csv("../../Resources/data/FacilityAddresses_API_v1.csv", encoding= 'unicode_escape')
```

```python
#Checking columns
campgrounds_address.columns
```

```python
#Dropping unnecessary columns 
campgrounds_postal = campgrounds_address.drop(columns=['FacilityAddressID', 'FacilityAddressType',
       'FacilityStreetAddress1', 'FacilityStreetAddress2',
       'FacilityStreetAddress3', 'City', 
       'AddressCountryCode', 'LastUpdatedDate'])
campgrounds_postal
```

```python
#Merging
data_West_Coast_campgrounds = data_West_Coast_campgrounds.merge(campgrounds_postal, on="FacilityID", how='inner')
data_West_Coast_campgrounds
```

```python
#Checking that only West Coast states include in the final dataset
data_West_Coast_campgrounds.groupby("PostalCode").size()
```

```python
#Renaming states in order to clean the data
data_West_Coast_campgrounds["PostalCode"].mask(data_West_Coast_campgrounds["PostalCode"] == "California", "CA", inplace=True)
data_West_Coast_campgrounds["PostalCode"].mask(data_West_Coast_campgrounds["PostalCode"] == "OREGON", "OR", inplace=True)
data_West_Coast_campgrounds["PostalCode"].mask(data_West_Coast_campgrounds["PostalCode"] == "Oregon", "OR", inplace=True)
data_West_Coast_campgrounds["PostalCode"].mask(data_West_Coast_campgrounds["PostalCode"] == "WASHINGTON", "WA", inplace=True)
data_West_Coast_campgrounds["PostalCode"].mask(data_West_Coast_campgrounds["PostalCode"] == "Washington", "WA", inplace=True)
```

```python
data_West_Coast_campgrounds.groupby("PostalCode").size()
```

```python
#Extracting only West Coast data
data_West_Coast_campgrounds = data_West_Coast_campgrounds.loc[
                                            (data_West_Coast_campgrounds["PostalCode"] == "CA") |
                                            (data_West_Coast_campgrounds["PostalCode"] == "OR") |
                                            (data_West_Coast_campgrounds["PostalCode"] == "WA") 
]
data_West_Coast_campgrounds
```

```python
#Checking on duplicates
data_West_Coast_campgrounds["FacilityName"].nunique()
```

```python
#Extracting duplicates
duplicate = data_West_Coast_campgrounds[data_West_Coast_campgrounds.duplicated(subset=["FacilityName"], keep=False)]
duplicate
```

```python
#Dropping the duplicates
data_West_Coast_campgrounds = data_West_Coast_campgrounds.drop_duplicates(subset=['FacilityID'])
data_West_Coast_campgrounds
```

```python
#Storing the full West Coat dataset into csv file
data_West_Coast_campgrounds.to_csv("../../Resources/data/West_Coast_activities_facilities_reservations.csv", index = False, header=True)
```
