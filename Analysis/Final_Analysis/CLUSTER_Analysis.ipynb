```python
import pandas as pd
```

```python
demand_time = pd.read_csv("../../Resources/data/Summary_statistics_for_ML.csv", encoding = 'unicode_escape')
demand_time
```

```python
WestCoast_facilities_CLUSTER = pd.read_csv("../../Resources/data/West_Coast_activities_facilities_reservations_CLUSTER.csv", encoding = 'unicode_escape')
WestCoast_facilities_CLUSTER
```

```python
dummy = WestCoast_facilities_CLUSTER["ActivityName"].str.get_dummies(sep=',')
dummy["FacilityID"] = WestCoast_facilities_CLUSTER["FacilityID"]
dummy
```

```python
activities_analysis = WestCoast_facilities_CLUSTER.merge(demand_time, left_on="FacilityID", right_on="facilityid", how="inner")
activities_analysis
```

```python
activities_analysis = activities_analysis.merge(dummy, on="FacilityID", how="inner")

activities_analysis
```

```python
activities_analysis["NumberofActivities"] = activities_analysis.values[:,21:122].sum(axis =1)
activities_analysis
```

```python
activities_analysis.loc[activities_analysis["Cluster"] == 5]
```

```python
activities_analysis.dtypes
```

```python
activities_analysis["NumberofActivities"] = pd.to_numeric(activities_analysis["NumberofActivities"])
```

```python
activities_analysis.dtypes
```

```python
cluster_analysis = pd.DataFrame(activities_analysis.groupby("Cluster").size())
cluster_analysis
```

```python
cluster_analysis = cluster_analysis.rename(columns={0: "number_of_camgrounds"})
```

```python
cluster_analysis["average_number_campsites"] = activities_analysis.groupby("Cluster")["NumberofCampsites"].mean()
cluster_analysis["median_number_campsites"] = activities_analysis.groupby("Cluster")["NumberofCampsites"].median()
```

```python
cluster_analysis["min_demand"] = activities_analysis.groupby("Cluster")["demand"].min()
cluster_analysis["mean_demand"] = activities_analysis.groupby("Cluster")["demand"].mean()
cluster_analysis["median_demand"] = activities_analysis.groupby("Cluster")["demand"].median()
cluster_analysis["max_demand"] = activities_analysis.groupby("Cluster")["demand"].max()
cluster_analysis
```

```python
cluster_analysis["average_median_orderdate_to_startdate"] = activities_analysis.groupby("Cluster")["median_orderdate_to_startdate"].mean()
cluster_analysis["median_median_orderdate_to_startdate"] = activities_analysis.groupby("Cluster")["median_orderdate_to_startdate"].median()
cluster_analysis["average_min_orderdate_to_startdate"] = activities_analysis.groupby("Cluster")["min_orderdate_to_startdate"].mean()
cluster_analysis["average_nights"] = activities_analysis.groupby("Cluster")["average_nights"].mean()
cluster_analysis["total_number_activities"] = activities_analysis.groupby("Cluster")["NumberofActivities"].sum()
cluster_analysis["average_number_activities"] = activities_analysis.groupby("Cluster")["NumberofActivities"].mean()
cluster_analysis["median_number_activities"] = activities_analysis.groupby("Cluster")["NumberofActivities"].median()
cluster_analysis
```

```python
cluster_analysis.to_csv("../../Resources/data/Analysis_each_CLUSTER.csv", index = False, header=True)
```

```python
cluster_activities = activities_analysis.groupby("Cluster").sum()
cluster_activities
```
