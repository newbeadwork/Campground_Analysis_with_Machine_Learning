```python
import pandas as pd
import datetime as dt
```

```python
reservations_WestCoast = pd.read_csv("reservations_West_Coast.csv", encoding = 'unicode_escape')
```

```python
reservations_WestCoast.dtypes
```

```python
reservations_WestCoast["startdate"] = pd.to_datetime(reservations_WestCoast["startdate"])
reservations_WestCoast["enddate"] = pd.to_datetime(reservations_WestCoast["enddate"])
reservations_WestCoast["orderdate"] = pd.to_datetime(reservations_WestCoast["orderdate"])
```

```python
reservations_WestCoast.dtypes
```

```python
reservations_WestCoast["orderdate_to_startdate"] = (reservations_WestCoast["startdate"] - reservations_WestCoast["orderdate"]).dt.days
```

```python
reservations_WestCoast["nights"] = (reservations_WestCoast["enddate"] - reservations_WestCoast["startdate"]).dt.days
```

```python
reservations_WestCoast.dtypes
```

```python
reservations_WestCoast
```

```python
minus = reservations_WestCoast.loc[reservations_WestCoast["orderdate_to_startdate"] < 0]
minus
```

```python
reservations_WestCoast = reservations_WestCoast.loc[reservations_WestCoast["orderdate_to_startdate"] >= 0]
reservations_WestCoast
```

```python
reservations_WestCoast["orderdate_to_startdate"].min()
```

```python
demand_time = reservations_WestCoast.groupby("facilityid")["orderdate_to_startdate"].median()
demand_time = pd.DataFrame(demand_time)
demand_time = demand_time.rename(columns={"orderdate_to_startdate": "median_orderdate_to_startdate"})
demand_time
```

```python
demand_time["average_orderdate_to_startdate"] = reservations_WestCoast.groupby("facilityid")["orderdate_to_startdate"].mean()
demand_time
```

```python
demand_time["min_orderdate_to_startdate"] = reservations_WestCoast.groupby("facilityid")["orderdate_to_startdate"].min()
demand_time
```

```python
demand_time["max_orderdate_to_startdate"] = reservations_WestCoast.groupby("facilityid")["orderdate_to_startdate"].max()
demand_time
```

```python
demand_time["average_nights"] = reservations_WestCoast.groupby("facilityid")["nights"].mean()
demand_time
```

```python
demand_time["median_nights"] = reservations_WestCoast.groupby("facilityid")["nights"].median()
demand_time
```

```python
demand_time.dtypes
```

```python
demand_time = demand_time.reset_index()
demand_time
```

```python
demand_time.to_csv("../../Resources/data/Summary_statistics_for_ML.csv", index = False, header=True)
```
